```python
import pandas as pd
import numpy as np
```

```python
students = pd.read_excel('Students28.xlsx')
students.tail()
```

```python
#追加列
students['Age'] = 25
students.tail()
```

```python
#删除列.drop
students.drop(columns=['Score', 'Age'], inplace= True)
students.tail()
```

```python
#插入列.insert
students.insert(1, column='Foo', value= np.repeat('foo', len(students)))
students.tail()
```

```python
#改列名.rename
students.rename(columns={'Foo': 'FOO', 'Name':'NAME'}, inplace= True)
students.tail()
```

```python
#设置空值
students['ID']= students['ID'].astype(float)
for i in range(8, 17):
    students['ID'].at[i] = np.nan
students.tail()
```

```python
#去掉空值.dropna
students.dropna(inplace=True)
students.tail()
```
