```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
homes = pd.read_excel('home_data.xlsx')
print(homes.head())
```

```python
#散点图
homes.plot.scatter(y= 'price', x= 'sqft_living')
#homes.plot.scatter(x= 'price', y= 'sqft_living')
#ax= plt.gca()
#ax.set_xticklabels(homes['price'], rotation= 45, ha= 'right')
plt.show()
```

```python
#直方图
homes.price.plot.hist(bins= 110)#bins设置成100个桶，会显示精细
plt.xticks(range(0, max(homes.price), 400000), fontsize= 8, rotation= 90)
plt.show()
```

```python
#直方图
homes.sqft_living.plot.hist(bins= 100)#bins设置成100个桶，会显示精细
plt.xticks(range(0, max(homes.sqft_living), 1000), fontsize= 8, rotation= 90)
plt.show()
```
