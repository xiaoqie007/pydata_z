```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
students = pd.read_excel('Students010.xlsx')
students.sort_values(by= '2017', inplace= True, ascending= False)
#students = students.sort_values(by= '2017', ascending= False)
print(students)
```

```python
students.plot.bar(x= 'Field', y= ['2016', '2017'], color= ['orange', 'red'], title= 'International Students by field')
#plt.tight_layout() #紧凑型
#plt.show()
```

```python
students.plot.bar(x= 'Field', y= ['2016', '2017'], color= ['blue', 'orange'])
#plt.tight_layout() #紧凑型
plt.title('International Students by field', fontsize= 15, fontweight= 'bold')
plt.xlabel('Field', fontweight= 'bold')
plt.ylabel('Number', fontweight= 'bold')
ax= plt.gca()
ax.set_xticklabels(students['Field'], rotation= 45, ha= 'right')
f =plt.gcf()
f.subplots_adjust(left= 0.2, bottom= 0.4)
plt.show()
```
