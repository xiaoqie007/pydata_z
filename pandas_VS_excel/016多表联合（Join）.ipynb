```python
#join方法

#table= students.join(scores, how= 'left').fillna(0)
table= students.join(scores, how= 'left', on= 'ID').fillna(0)
#注意在.join中 已经没有left_on，right_on了。
table.Score= table.Score.astype(int)
print(table)
```
