```python
import pandas as pd
```

```python
#低于60分的红色字体，其于的黑色字体
def low_score_red(s):
    color= 'red' if s < 60 else 'black'
    return f'color:{color}'
#每次成绩的第一名
def hightest_score_green(col):
    return['background-color:lime' if s == col.max() else 'background-color: white' for s in col ]
```

```python
students= pd.read_excel('Students25.xlsx')
students.style.applymap(low_score_red, subset= ['Test_1', 'Test_2', 'Test_3']) \
.apply(hightest_score_green, subset= ['Test_1', 'Test_2', 'Test_3'])  #加了一个折行和一个反斜杠
```
