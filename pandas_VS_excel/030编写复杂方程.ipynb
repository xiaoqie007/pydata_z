```python
import pandas as pd
import numpy as np
```

```python
def get_circumcircle_area(l, h):
    r = np.sqrt(l**2 + h**2)/2
    return r**2*np.pi
```

```python
def wrapper(row):
    return get_circumcircle_area(row['Length'], row['Height'])
```

```python
rects = pd.read_excel('Rectangles.xlsx', index_col= 'ID')
rects.head()
rects['Circumcirle Area'] = rects.apply(wrapper, axis= 1)
rects
```
