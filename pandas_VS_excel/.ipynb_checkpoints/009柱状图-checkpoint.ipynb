```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
students = pd.read_excel('Students09.xlsx')
students = students.sort_values(by= 'Number', ascending= False)
print(students)
```

```python
students.plot.bar(x= 'Field', y= 'Number', color= 'red', title= 'International Students by field')
#plt.tight_layout() #紧凑型
plt.show()
```

```python
plt.bar(students.Field, students.Number, color= 'blue')
plt.xticks(students.Field, rotation = '90')
plt.xlabel('Field', fontsize = 13)
plt.ylabel('Number', fontsize= 13)
plt.title('International Students by field', fontsize = 15)
plt.show()
```
