```python
#注意读取Excel文件中的2016，2017是字符串，即‘2016， ’2017.
```

```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
students = pd.read_excel('Students12.xlsx', index_col= 'From')
#print(students)
#students['2017'].plot.pie(fontsize= 5)
students['2017'].plot.pie(fontsize= 5,counterclock= False)#counterclock把图倒过来
#students['2017'].sort_values(ascending= True).plot.pie(fontsize= 5, startangle=340)
#利用sort_values排序的方法把图倒过来
plt.title('Source of International Students', fontsize= 13, fontweight= 'bold')
plt.ylabel('2017', fontsize= 10, fontweight= 'bold')
plt.show()
```

```python
students = pd.read_excel('Students12.xlsx', index_col= 'From')
#print(students)
students['2016'].sort_values(ascending= True).plot.pie(fontsize= 5)
#students['2016'].sort_values(ascending= True).plot.pie(fontsize= 5, startangle=340)
plt.title('Source of International Students', fontsize= 13, fontweight= 'bold')
plt.ylabel('2016', fontsize= 10, fontweight= 'bold')
plt.show()
```
