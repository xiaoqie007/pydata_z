```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
#首先打开后发现‘Month’类型#float64
sales= pd.read_excel('Sales.xlsx')
print(sales)
print(sales['Month'].dtype) #float64
#于是使用dtype= {'Month':str}，将‘Month’设置成str
sales= pd.read_excel('Sales.xlsx', dtype= {'Month':str})
print(sales)
print(sales['Month'].dtype)
```

```python
plt.bar(sales.Month, sales.Revenue)
plt.title('Sales', fontsize= 14, fontweight= 'bold')
plt.xticks(sales.Month, rotation= 80, fontsize= 9)
plt.show()
```

```python
#按照课程里面的实例加上sales.index

plt.bar(sales.index, sales.Revenue)
plt.title('Sales', fontsize= 14, fontweight= 'bold')
plt.xticks(sales.index, sales.Month, rotation= 80, fontsize= 9)
#plt.tight_layout()
plt.show()
```

```python
#线性回归，数据预测
import pandas as pd
import matplotlib.pylab as plt
from scipy.stats import linregress #linregress线性回归单词缩写
```

```python

sales= pd.read_excel('Sales.xlsx', dtype= {'Month':str})

slope,intercept, r, p,std_err= linregress(sales.index, sales.Revenue)#调取线性回归的函数取值

exp= sales.index * slope + intercept #生成期望值表达式

plt.scatter(sales.index, sales.Revenue)#原来的bar改成scatter

plt.plot(sales.index, exp, color= 'orange')#生成期望值图像

plt.title('Sales', fontsize= 14, fontweight= 'bold')

plt.title(f'y= {slope} * x + {intercept}')#显示期望值表达式

plt.xticks(sales.index, sales.Month, rotation= 90, fontsize= 9)
#plt.tight_layout()
plt.show()

# 预测2019年12月的销售额
print(slope*35+intercept)
```
