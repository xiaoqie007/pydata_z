```python
import pandas as pd
import seaborn as sns
```

```python
#分数越高，颜色越深
color_map = sns.light_palette('green', as_cmap= True) #light_palette是sns里面的调色板

students= pd.read_excel('Students25.xlsx', index_col='ID')
students.style.background_gradient(color_map, subset= ['Test_1', 'Test_2','Test_3'])#color_map映射到后面的三列上
```

```python
#分数越高，数据条越长
import pandas as pd
students= pd.read_excel('Students25.xlsx', index_col= 'ID')
students.style.bar(color= 'red', subset= ['Test_1', 'Test_2','Test_3']) # 使用bar
```
